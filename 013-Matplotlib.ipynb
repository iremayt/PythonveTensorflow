```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g") #plt.plot(x, y, "g")-->g:green,r:red,b:blue,y:yellow...
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Yaş-Kilo Grafik Başlığı")
#plt.show()
```

## ozellestirme:

```python
numpyDizi1 = np.linspace(0, 10, 20) #.linspace():0 ile 10 arasinda, 10 dahil, esit aralikli 20 sayi
numpyDizi2 = numpyDizi1 ** 3
```

```python
plt.plot(numpyDizi1, numpyDizi2, "r--") #"r--"
```

```python
plt.plot(numpyDizi1,numpyDizi2,"r*-") #"r*-"
```

```python
plt.plot(numpyDizi1,numpyDizi2,"r+-") #"r+-"
```

## subplot:

subplot(): Hızlı alt grafikler oluşturmak için kullanılabilir. İlk iki parametreye oluşturulacak satır ve sütun sayısı girilir,

```python
plt.subplot(1, 2, 1) #1 sira, 2 kolon, 1.yi ciziyorum
plt.plot(numpyDizi1, numpyDizi2, "y*-")

plt.subplot(1, 2, 2) #1 sira, 2 kolon, 2.yi ciziyorum
plt.plot(numpyDizi2, numpyDizi1, "b:")
```

## figure:

Basit bir figür(alan) oluşturmak için plt.figure(), çevresine eksen eklemek için ise plt.axes() kullanmaktayız.

```python
myFigure1 = plt.figure()
figureAxes = myFigure1.add_axes([0.4, 0.4, 0.6, 0.6]) #.add_axes([x baslangic uzakligi, y baslangic uzakligi, width, height])

figureAxes.plot(numpyDizi1, numpyDizi2, "g")
figureAxes.set_xlabel("X ekseni ismi")
figureAxes.set_ylabel("Y ekseni ismi")
figureAxes.set_title("Grafik Başlığı")
```

### subplots vs. figure:

```python
plt.subplots()
#aslinda bir tuple dondurur.
#sadece figure dgl 1 eksende cikarir (Figure with 1 axes)
#Dolayisiyla figure den biraz daha farkli: hem figure hem axes var.
```

```python
plt.figure()
#0 eksenli bir figure olusur. (Figure with 0 axes)
```

```python
#subplots tuple dondurdugu icin, o zmn bir tuple a esitleyebilirim:
(benimFigure, benimEksen) = plt.subplots() #(1.alan figure dondurur, 2.alan eksen dondurur):(benimFigure, benimEksen)
```

```python
benimFigure #figure u cagirirsam, bos figure gelir
```

```python
benimEksen #eksen i cagirirsam, eksen gelir
```

## figure kullanarak içiçe grafikler:

```python
myFigure2 = plt.figure()
#Ayni figure icinde 2 farkli eksen kullanarak yaptik:
eksen1 = myFigure2.add_axes([0.1, 0.1, 0.9, 0.9])
eksen2 = myFigure2.add_axes([0.2, 0.5, 0.3, 0.3]) #kucuk grafigin sol alt kose koordinati:(x=0.2, y=0.5)

eksen1.plot(numpyDizi1, numpyDizi2, "r")
eksen1.set_xlabel("X EKSENİ")
eksen1.set_ylabel("Y EKSENİ")
eksen1.set_title("ANA GRAFİK BAŞLIK")

eksen2.plot(numpyDizi2, numpyDizi1, "b")
eksen2.set_xlabel("x ekseni küçük")
eksen2.set_ylabel("y ekseni küçük")
eksen2.set_title("küçük grafik başlık")
```

## subplot kullanarak içiçe grafikler:

```python
(myFigure, myEksenler) = plt.subplots(nrows= 1, ncols= 3)

for eksen in myEksenler:
    eksen.plot(numpyDizi1, numpyDizi2, "g")
    eksen.set_xlabel("x ekseni")
    
plt.tight_layout()
```

# görsel geliştirmeler

```python
gFigure=plt.figure(dpi=100)
gEksen=gFigure.add_axes([0.1,0.1,0.9,0.9])

gEksen.plot(numpyDizi1,numpyDizi2**2,label="numpyDizi1'in karesi")
gEksen.plot(numpyDizi1,numpyDizi2**3,label="numpyDizi1'in küpü")

gEksen.legend(loc=6) #loc=1,2,3,4,...,10...
```

```python
gFigure.savefig("gorselfigure.png",dpi=200) #kaydetme islemi
```
